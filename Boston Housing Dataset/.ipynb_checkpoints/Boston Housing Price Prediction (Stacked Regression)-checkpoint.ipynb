### Prediction Modelling

```python
# We will now get dummies for all the categorical features using LabelEncoder. 
# We will prepare the data first below for this

ntrain = dfTrain.shape[0]
ntest = dfTest.shape[0]
y_train = dfTrain.SalePrice.values

all_data = pd.concat((dfTrain, dfTest)).reset_index(drop=True)
all_data.drop(['SalePrice'], axis=1, inplace=True)

print("Full data size is : {}".format(all_data.shape))
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ('FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 
        'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 'KitchenQual', 'BsmtFinType1', 
        'BsmtFinType2', 'Functional', 'Fence', 'BsmtExposure', 'GarageFinish', 'LandSlope',
        'LotShape', 'PavedDrive', 'Street', 'Alley', 'CentralAir', 'MSSubClass', 'OverallCond', 
        'MoSold')

# apply LabelEncoder to categorical features above
for c in cols:
    lbl = LabelEncoder() 
    lbl.fit(list(all_data[c].values)) 
    all_data[c] = lbl.transform(list(all_data[c].values))

# shape of the new data       
print('Shape all_data: {}'.format(all_data.shape))
```

```python
# Get dummies
all_data = pd.get_dummies(all_data)
print(all_data.shape)
```

```python
# Split into train and test again
train = all_data[:ntrain]
test = all_data[ntrain:]
```

```python
# Imports
from sklearn.linear_model import ElasticNet, Lasso  
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#Validation function using cross_val_score

n_folds = 5

def rmsle_cv(model):
    kf = KFold(n_folds, shuffle=True, random_state=42).get_n_splits(train.values)
    rmse= np.sqrt(-cross_val_score(model, train.values, y_train, scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

#### Base models scores

```python
#LASSO & ELASTICNET : using robustscaler in pipeline

lasso = make_pipeline(RobustScaler(), Lasso(alpha =0.0005, random_state=1))

ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9, random_state=3))


GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.05,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
```

```python
# Run lasso

score = rmsle_cv(lasso)
print("Lasso score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
# Run ENet

score = rmsle_cv(ENet)
print("ElasticNet score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))

```

```python
# Run GBoost

score = rmsle_cv(GBoost)
print("Gradient Boosting score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

#### Averaging function for stacking mulitple models

```python
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, models):
        self.models = models
        
    # we define clones of the original models to fit the data in
    def fit(self, X, y):
        self.models_ = [clone(x) for x in self.models]
        
        # Train cloned base models
        for model in self.models_:
            model.fit(X, y)

        return self
    
    #Now we do the predictions for cloned models and average them
    def predict(self, X):
        predictions = np.column_stack([
            model.predict(X) for model in self.models_
        ])
        return np.mean(predictions, axis=1)   
```

```python
# Run with all three models

averaged_models = AveragingModels(models = (ENet, GBoost, lasso))

score = rmsle_cv(averaged_models)
print("Averaged base models score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

```python
# Validation function
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
```

```python
# Fit and predict
averaged_models.fit(train.values, y_train)
train_pred = averaged_models.predict(train.values)
test_pred = np.expm1(averaged_models.predict(test.values))
print(rmsle(y_train, train_pred))
```
